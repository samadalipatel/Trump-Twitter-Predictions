The following loop will take sets of 100 ids (since the API has a 100-id-limit) and search for tweets that have been deleted. 

```python
# Create empty list to append output ids to - it's called semifinal since it will be a list of integers 
semifinal_list = []

# Create loop to append ids to 
for i in np.arange(0, len(df), 100):
    # Create ids
    input_ids = list(df.id_str[i:i+100].values)
    
    # Look up each tweet based on id 
    tweets = api.statuses_lookup(id_= input_ids)
    print('There are', len(tweets), 'remaining from this set.')
    
    # Extract the tweets into a list 
    output_list = [tweet.id for tweet in tweets]
    
    # Extend final_list with output_list
    semifinal_list.extend(output_list)

# Change datatype of semi_final list to string as id_str is a string 
final_list = [str(item) for item in semifinal_list]
```

```python
# Save the non-deleted tweets into a df
df2 = df[df.id_str.isin(final_list)]
```
